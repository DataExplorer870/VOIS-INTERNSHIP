```python
plt.figure(figsize=(10,10))
plt.scatter(x='income', y='visits', data=df)
plt.xlabel('Income')
plt.ylabel('Visits')
plt.title('Income vs Visits')
```

```python
sns.histplot(df.gender, bins=2)
```

```python
label=['yes','no']
Y =df[df['freepoor']=='yes']
N =df[df['freepoor']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label,autopct='%1.1f%%')
plt.title('% of people who are getting govt health insurance due to low income')
plt.show()
Y =df[df['private']=='yes']
N =df[df['private']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label,autopct='%1.1f%%')
plt.title('% of people who are having private health insurance')
plt.show()
Y =df[df['freerepat']=='yes']
N =df[df['freerepat']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label,autopct='%1.1f%%')
plt.title('% of people who are getting govt health insurance due to old age, disability or veteran status')
plt.show()
```

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['crimson','lightcoral'])
plt.title('bar chart')
plt.xlabel('reduced activity')
plt.ylabel('gender')
plt.show()
```

```python
db =df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.bar(db['gender'], db['reduced'], color=['blue', 'skyblue'])
plt.title('Reduced Activity by Gender')
plt.xlabel('Gender')
plt.ylabel('Reduced Activity')
plt.show()
```
